```python
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
condition = school_data_complete["reading_score"] >= 70
school_data_complete[condition]["student_name"].count() / school_data_complete["student_name"].count()
```

```python
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
total_budget = school_data_complete[["school_name", "budget"]]["budget"].unique().sum()
average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()
percent_math = (school_data_complete[school_data_complete["math_score"] >= 70]["student_name"].count() / school_data_complete["student_name"].count()) * 100
percent_reading = (school_data_complete[school_data_complete["reading_score"] >= 70]["student_name"].count() / school_data_complete["student_name"].count()) * 100
overall_passing = (school_data_complete["math_score"].mean() + school_data_complete["reading_score"].mean()) / 2


results_df = pd.DataFrame({"Total Schools": total_schools,
                          "Total Students": '{:0,}'.format(total_students),
                          "Total Budget": '${:0,.2f}'.format(total_budget),
                          "Average Math Score": average_math,
                          "Average Reading Score": average_reading,
                          "% Passing Math": percent_math,
                          "% Passing Reading": percent_reading,
                          "% Overall Passing Rate": overall_passing}, index = [0])

results_df
```

```python
labels = ["Total Schools", "Total Students", "Total Budget", "Average Math Schore", "Average Reading Score",
         "% Passing Math", "% Passing Reading", "% Overall Passing Rage"]
values = [total_schools, total_students, '${:0,.2f}'.format(total_budget), average_math, average_reading, 
         percent_math, percent_reading, overall_passing]
results_dict = dict(zip(labels, values))
results_dict
```

```python
results_zip_df = pd.DataFrame(results_dict, index=[0])
results_zip_df
```

```python
passing_math = school_data_complete["math_score"] >= 70
passing_reading = school_data_complete["reading_score"] >= 70
school_data_complete["Passing_Math"] = passing_math
school_data_complete["Passing_Reading"] = passing_reading
school_data_complete.head()
```

```python
school_group2 = school_data_complete.groupby(["school_name"], as_index=True).agg({'type':'first', 'student_name':'count',
                                                                  'budget':'first', 'math_score':'mean',
                                                                  'reading_score':'mean', 'Passing_Math':'sum',
                                                                  'Passing_Reading':'sum'})
school_group2.index.name=None
school_group2
```

```python
# school_group2_df = school_group2.reset_index()
school_group2["% Passing Math"] = school_group2["Passing_Math"] / school_group2["student_name"]
school_group2["% Passing Reading"] = school_group2["Passing_Reading"] / school_group2["student_name"]
school_group2["Overall Passing Rate"] = (school_group2["% Passing Math"] + school_group2["% Passing Reading"]) / 2
school_group2.head()
```

```python
school_group2["Per Student Budget"] = school_group2["budget"] / school_group2["student_name"]
school_group2.head()
```

```python
# need to reformat some values, probably

school_summary = pd.DataFrame({
#     "School Name": school_group2["school_name"],
    "School Type": school_group2["type"],
    "Total Students": school_group2["student_name"],
    "Total School Budget": school_group2["budget"],
    "Per Student Budget": school_group2["Per Student Budget"],
    "Average Math Score": school_group2["math_score"],
    "Average Reading Score": school_group2["reading_score"],
    "% Passing Math": school_group2["% Passing Math"],
    "% Passing Reading": school_group2["% Passing Reading"],
    "Overall Passing Rate": school_group2["Overall Passing Rate"]
})
school_summary
```
